Comparação de ETF com Índice em Busca de Minimizar o Tracking Error

```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
import plotly.graph_objects as go
```

```python
# Download das cotações históricas do OEF e SP100
returns = pd.DataFrame(yf.download('^SP100','2016-12-29',datetime.date.today()+datetime.timedelta(1))['Adj Close'])
returns['OEF'] = yf.download('OEF','2016-12-29',datetime.date.today()+datetime.timedelta(1))['Adj Close']

# Cálculo da variação % diária dos ativos
returns = returns.apply(lambda x: (x / x.shift(1)-1).fillna(0))

# Renomeando as colunas
returns.columns = ['SP100','OEF']

# Cálculo do retorno acumulado para o período selecionado
returns = ((1 + returns).cumprod())*100
returns
```

```python
fig = go.Figure(data=go.Scatter(x=returns.index,y=returns['SP100'],name='SP100'))
fig.add_trace(go.Scatter(x=returns.index,y=returns['OEF'],name='OEF'))
fig.show()
```
