```python
# Importação das bibliotecas a serem utilizadas
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
import plotly.express as px
```

Ações componentes do S&P100

```python
# Criação de dataframe vazio para posterior iteração
componentes_sp100 = pd.DataFrame()

# Iteração para adicionar uma coluna ao dataframe contendo a composição do S&P100 com base na planilha em excel
for i in pd.ExcelFile('Composições_SP100.xlsx').sheet_names:
    componentes_sp100[i] = pd.read_excel('Composições_SP100.xlsx',sheet_name=i)['Código']

componentes_sp100
```

```python
componentes_sp100.info()
```

```python
# Criação de lista vazia para posterior iteração
ativos = []

# Iteração para pegar cada coluna dos componentes e adicionar em uma lista
for i in componentes_sp100.columns:
    ativos += list(componentes_sp100[i].drop_duplicates().dropna())
    
ativos
```

Retorno das Ações Componentes do S&P100

```python
# Fazer o download das cotações históricas nas datas solicitadas
returns = yf.download(ativos,'2016-06-29','2024-04-16')['Adj Close']

# Fazer o cálculo de variação % diária com base na coluna 'Adj Close'
returns = returns.apply(lambda x: (x / x.shift(1) - 1).fillna(0))

# Visualização da tabela contendo retornos diários de todas as ações
returns
```

```python
returns.info()
```

Retorno do S&P100

```python
# Download das cotações históricas do S&P100
sp100_returns = pd.DataFrame(yf.download('^SP100','2016-06-29','2024-04-16')['Adj Close'])

# Fazer o cálculo de variação % diária com base na coluna 'Adj Close'
sp100_returns = sp100_returns.apply(lambda x: (x / x.shift(1) - 1).fillna(0))
sp100_returns.columns = ['S&P100']
sp100_returns
```

```python
px.line((1 + sp100_returns).cumprod())
```
