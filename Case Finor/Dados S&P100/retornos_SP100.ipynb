```python
# Importação das bibliotecas a serem utilizadas
import pandas as pd
import numpy as np
import yfinance as yf
```

Retorno das Ações componentes do SP100

```python
# Ler arquivo de composição do S&P100
sp100_comp = pd.read_excel('SP100_Wiki.xlsx')
sp100_comp
```

```python
# Criação de uma lista contendo todas as ações listadas do índice
tickers = list(sp100_comp['Código'])
tickers
```

```python
# Fazer o download das cotações históricas nas datas solicitadas
returns = yf.download(tickers,'2023-12-28','2024-04-06')['Adj Close']

# Fazer o cálculo de variação % diária com base na coluna 'Adj Close'
returns = returns.apply(lambda x: (x / x.shift(1) - 1).fillna(0))
```

```python
# Visualização da tabela contendo retornos diários de todas as ações
returns.head()
```
