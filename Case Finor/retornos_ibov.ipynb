```python
# Importação das bibliotecas a serem utilizadas
import pandas as pd
import numpy as np
import yfinance as yf
```

Retorno das Ações componentes do IBOV

```python
# Ler arquivo de composição do IBOVESPA retirado do site da B3
ibov_comp = pd.read_excel('IBOVDia_08-04-24.xlsx')

# Adicionar aos códigos '.SA' para ficar no padrão do YFINANCE
ibov_comp['Código'] = ibov_comp['Código'].apply(lambda x: str(x + '.SA')) 

# Ordenar as ações com maior peso no índice (apenas para visualizar)
ibov_comp.sort_values(by='Part. (%)',ascending=False)
```

```python
# Criação de uma lista contendo todas as ações listadas do índice
tickers = list(ibov_comp['Código'])
tickers
```

```python
# Fazer o download das cotações históricas nas datas solicitadas
returns = yf.download(tickers,'2023-12-28','2024-04-11')['Adj Close']

# Fazer o cálculo de variação % diária com base na coluna 'Adj Close'
returns = returns.apply(lambda x: (x / x.shift(1) - 1).fillna(0))
```

```python
# Visualização da tabela contendo retornos diários de todas as ações
returns.head()
```

Retorno do IBOV
