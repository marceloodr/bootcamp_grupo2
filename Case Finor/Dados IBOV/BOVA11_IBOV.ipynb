Comparação de ETF com Índice em Busca de Minimizar o Tracking Error

```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
import plotly.graph_objects as go
```

```python
# Download das cotações históricas do IBOV e BOVA11
returns = pd.DataFrame(yf.download(['^BVSP','PETR4.SA','VALE3.SA','ITUB4.SA'],'2016-12-29',datetime.date.today()+datetime.timedelta(1))['Adj Close'])
# returns['BOVA11'] = yf.download('BOVA11.SA','2016-12-29',datetime.date.today()+datetime.timedelta(1))['Adj Close']

# Cálculo da variação % diária dos ativos
returns = returns.apply(lambda x: (x / x.shift(1)-1).fillna(0))
returns.columns = ['IBOV','PETR4','VALE3','ITUB4']

# Cálculo do retorno acumulado para o período selecionado
returns = ((1 + returns).cumprod())*100
returns
```

```python
# Criação do Gráfico

fig = go.Figure()

for i in returns.columns:
    fig.add_trace(go.Scatter(x=returns.index,y=returns[i],name=i))
fig.show()
```
