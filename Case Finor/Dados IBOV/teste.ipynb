```python
import time
import datetime
import pandas as pd
import numpy as np
import yfinance as yf
from scipy.optimize import minimize
import plotly.express as px
```

```python
today = datetime.date.today()
days_delta = datetime.timedelta(62)
end = datetime.date(2022,12,31)
start =  end - days_delta
```

```python
tickers = list(pd.read_excel('Composições_IBOV.xlsx',sheet_name='jan2023')['Código'].apply(lambda x: str(x + '.SA')))
tickers
```

```python
returns = yf.download(tickers,start,end)['Adj Close']
```

```python
returns = returns.apply(lambda x: (x / x.shift(1) - 1).fillna(0))
returns
```

```python
ibov_return = pd.DataFrame(yf.download('^BVSP',start,end)['Adj Close'])
ibov_return = ibov_return.apply(lambda x: (x / x.shift(1) - 1).fillna(0))
ibov_return.columns = ['IBOV']
ibov_return
```

```python
corr_df = returns
corr_df['IBOV'] = ibov_return['IBOV']
corr_df = corr_df.corr()
corr_df
```

```python
n_ativos = 10

ativos = corr_df['IBOV'].sort_values(ascending=False).drop('IBOV').head(n_ativos)
ativos
```

```python
ativos_returns =returns[list(ativos.index)]*100
ativos_returns
```

```python
def objective(weights, returns_assets, returns_ibovespa):
    portfolio_returns = np.dot(returns_assets, weights)
    return np.sum((portfolio_returns - returns_ibovespa) ** 2)

# Restrições: os pesos devem somar 1
constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})

# Limites de peso entre 0 e 1
bounds = [(0, 1)] * len(ativos_returns.columns)

# Chute inicial para os pesos
initial_guess = np.ones(len(ativos_returns.columns)) / len(ativos_returns.columns)

# Executa a otimização
result = minimize(objective, initial_guess, 
                  args=(ativos_returns.values, ibov_return.values),
                  method='SLSQP', constraints=constraints, bounds=bounds)

# Os pesos ótimos
optimal_weights = result.x
optimal_weights
```

```python
port_return = (ativos_returns * optimal_weights).sum(axis=1)
ibov_return = ibov_return * 100
ibov_return['Portfólio'] = port_return
ibov_return
```

```python
base = (1+(ibov_return/100)).cumprod()
px.line(base)
```

```python
teste_ibov = pd.DataFrame(yf.download('^BVSP',end,today)['Adj Close'])
teste_ibov = teste_ibov.apply(lambda x: (x /x.shift(1)-1).fillna(0))
teste_ibov
```

```python
teste_ativos = pd.DataFrame(yf.download(list(ativos.index),end,today)['Close'])
teste_ativos = teste_ativos.apply(lambda x: (x /x.shift(1)-1).fillna(0))
teste_ativos = teste_ativos
```

```python
teste_return = (teste_ativos * optimal_weights).sum(axis=1)
teste_ibov = teste_ibov
teste_ibov['Portfólio'] = teste_return
teste_ibov
```

```python
base_teste = (1+teste_ibov).cumprod()
px.line(base_teste)
```
