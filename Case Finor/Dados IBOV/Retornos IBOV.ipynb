```python
# Importação das bibliotecas a serem utilizadas
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
```

Ações componentes do IBOV

```python
# Ler arquivo com as Ações componentes do IBOV #

# Criação de dataframe vazio para posterior iteração
componentes_ibov = pd.DataFrame()

# Iteração para adicionar uma coluna ao dataframe contendo a composição do IBOV com base na planilha em excel
for i in pd.ExcelFile('Composições_IBOV.xlsx').sheet_names:
    componentes_ibov[i] = pd.read_excel('Composições_IBOV.xlsx',sheet_name=i)['Código']

componentes_ibov = componentes_ibov.apply(lambda x: (x + '.SA'))
componentes_ibov
```

```python
componentes_ibov.info()
```

```python
# Obter a lista de todos os tickers existentes na amostra #

# Criação de dataframe vazio para posterior iteração
ativos = pd.DataFrame()

# Iteração para pegar cada coluna dos componentes e adicionar em uma lista
for i in componentes_ibov.columns:
    ativos_add = componentes_ibov[i]
    ativos = pd.concat([ativos,ativos_add])

# Retirar duplicadas e NaN
ativos = ativos.drop_duplicates(ignore_index=True).dropna(ignore_index=True).set_index(0)
ativos = list(ativos.index)
ativos
```

Retorno das Ações Componentes do IBOV

```python
# Fazer o download das cotações históricas nas datas solicitadas
returns = yf.download(ativos,'2016-06-29',datetime.date.today()+datetime.timedelta(1))['Adj Close']

# Fazer o cálculo de variação % diária com base na coluna 'Adj Close'
returns = returns.apply(lambda x: (x / x.shift(1) - 1).fillna(0))

# Visualização da tabela contendo retornos diários de todas as ações
returns.tail()
```

Retorno do IBOV
