```python
# Importação das bibliotecas a serem utilizadas
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
import plotly.express as px
```

Ações componentes do IBOV

```python
# Criação de dataframe vazio para posterior iteração
componentes_ibov = pd.DataFrame(index=range(0,93))

# Iteração para adicionar uma coluna ao dataframe contendo a composição do IBOV com base na planilha em excel
for i in pd.ExcelFile('Composições_IBOV_Anual.xlsx').sheet_names:
    componentes_ibov[i] = pd.read_excel('Composições_IBOV_Anual.xlsx',sheet_name=i)['Código']

# Adição de '.SA' para que os tickers sigam o padrão do Yahoo Finance
componentes_ibov = componentes_ibov.apply(lambda x: x + '.SA')
componentes_ibov
```

```python
componentes_ibov.info()
```

```python
# Criação de lista vazia para posterior iteração
ativos = []

# Iteração para pegar cada coluna dos componentes e adicionar em uma lista
for i in componentes_ibov.columns:
    ativos += list(componentes_ibov[i].drop_duplicates().dropna())
    
ativos
```

Retorno das Ações Componentes do IBOV

```python
# Fazer o download das cotações históricas nas datas solicitadas
returns = yf.download(ativos,'2016-06-29','2024-04-16')['Adj Close']

# Fazer o cálculo de variação % diária com base na coluna 'Adj Close'
returns = returns.apply(lambda x: (x / x.shift(1) - 1).fillna(0))

# Visualização da tabela contendo retornos diários de todas as ações
returns
```

```python
ativos_faltantes = pd.read_excel('Ativos_Faltantes.xlsx',index_col='Unnamed: 0')
ativos_faltantes.fillna(0,inplace=True)

returns = returns.drop(columns=ativos_faltantes.columns)
returns = returns.join(ativos_faltantes)

returns 
```

```python
returns.info()
```

```python
returns.isna().sum().sum()
```

```python
data_teste = datetime.date(2023,12,31)
n_dias = datetime.timedelta(30)

ativo_teste = 'PETR4.SA'

returns.loc[data_teste-n_dias:data_teste,ativo_teste] * 100
```

Retorno do IBOV

```python
# Download das cotações históricas do IBOV
ibov_returns = pd.DataFrame(yf.download('^BVSP','2016-06-29','2024-04-16')['Adj Close'])

# Fazer o cálculo de variação % diária com base na coluna 'Adj Close'
ibov_returns = ibov_returns.apply(lambda x: (x / x.shift(1) - 1).fillna(0))
ibov_returns.columns = ['IBOV']
ibov_returns
```

```python
ibov_returns.mean() * 100
```

```python
px.line((1 + ibov_returns).cumprod())
```

```python
data_teste = datetime.date(2023,12,31)
n_dias = datetime.timedelta(30)

ibov_returns.loc[data_teste-n_dias:data_teste] * 100
```
