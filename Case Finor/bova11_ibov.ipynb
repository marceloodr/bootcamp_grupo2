```python
import pandas as pd
import numpy as np
import yfinance as yf
import plotly.graph_objects as go
```

```python
# Download das cotações históricas do IBOV e BOVA11
returns = pd.DataFrame(yf.download('^BVSP','2016-12-29','2024-04-10')['Adj Close'])
returns['BOVA11'] = yf.download('BOVA11.SA','2016-12-29','2024-04-10')['Adj Close']

# Cálculo da variação % diária dos ativos
returns = returns.apply(lambda x: (x / x.shift(1)-1).fillna(0))

# Renomeando as colunas
returns.columns = ['IBOV','BOVA11']

# Cálculo do retorno acumulado para o período selecionado
returns = ((1 + returns).cumprod())*100
returns
```

```python
fig = go.Figure(data=go.Scatter(x=returns.index,y=returns['IBOV'],name='IBOV'))
fig.add_trace(go.Scatter(x=returns.index,y=returns['BOVA11'],name='BOVA11'))
fig.show()
```

Cálculo de Drawdown

```python
drawdown = returns / returns.cummax() - 1
```

```python
fig_dd = go.Figure(data=go.Scatter(x=drawdown.index,y=drawdown['IBOV'],name='IBOV'))
fig_dd.add_trace(go.Scatter(x=drawdown.index,y=drawdown['BOVA11'],name='BOVA11'))
fig_dd.show()
```
